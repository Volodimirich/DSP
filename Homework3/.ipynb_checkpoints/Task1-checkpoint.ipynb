```python
from scipy import signal
import numpy as np
import matplotlib.pyplot as plt
```

```python
M = 150
am = 301
n = np.linspace(-M, M, am)
```

```python
def plot_graph(b, name = 'x', phase = False, graph_name = 'Frequency response'):
    w, h = signal.freqz(b)
    w_n = w/max(w)
    
    if phase:
        angles = np.unwrap(np.angle(h))
        plt.plot(w_n, angles, label=f'{name}')
        plt.ylabel("Phase (rad)")

    else:
        h_n = 20 * np.log10 (abs(h))
        plt.plot(w_n,h_n, label=f'{name}')
        plt.ylabel("Magnitude (db)")

    
    plt.legend()
    plt.xlabel(r'Normalized Frequency (x$\pi$rad/sample)')
    plt.title(f'{graph_name}')
```

```python
h_calc = (1-np.exp(-1/2-2/5*1j*np.pi*n))/(2*1j*np.pi*n + 5/2) + (-1+np.exp(-1/2+2/5*1j*np.pi*n))/(2*1j*np.pi*n - 5/2)
```

```python
win = signal.get_window('hamming', am)
B = h_calc * np.transpose(win)
```

```python
plot_graph(B, phase = False, graph_name = 'Magnitude response')
```

```python
plot_graph(B, phase = True, graph_name = 'Phase response' )
```
