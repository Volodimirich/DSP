### Problem №3
#### Compute the auto-correlation of the triangular pulse

```python
def triangle_wave(x = 0, c = 1, hc = 1, k = 1): #  with amplitude hc, width c, and slope hc/2c
    if abs(x) >= c/k:
        return 0
    else:
        sign = 1 if x < 0 else -1
        return  hc + (k * sign * x * hc )/ c 

x=np.linspace(-3,3,1000)
y=np.array([triangle_wave(t) for t in x])
plt.plot(x, y)
plt.title('Triangle pulse')
plt.show()

act = np.correlate(y, y, mode='same')

plt.plot(x, act)
plt.title('Auto-correlation of the triangular pulse')
plt.show()
```

### Solution 5b

```python
interval = [0, 8]
sample_freq = 6000



x = np.linspace(interval[0], interval[1], (interval[1] - interval[0]) * sample_freq)
sin_orig = np.sin(2*np.pi*x)
downsampling_rate = len(x)//(8 * 2.5) 


idx = np.arange(0, len(x), downsampling_rate + 100).astype(int)
z_downsampled = np.zeros_like(sin_orig)
z_downsampled[idx] = sin_orig[idx]


# Plot original and downsampled versions of our signal
plt.figure()
plt.plot(x, sin_orig)  # From the previous cell
plt.plot(x, z_downsampled) # All except z_downsampled[idx] are zeros

# Now, restore the signal using sinc functions
def sinc(x):
    return np.sinc(x / downsampling_rate)


z_restored = np.zeros_like(z_downsampled)
z_samples = z_downsampled[idx]
for i in range(len(x)):
    z_restored[i] += np.sum(z_samples * sinc(idx - i))

plt.plot(x, z_restored)
plt.xlim([0, 4]) # Zoom the X-axis in   
```
